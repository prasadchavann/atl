```python
#libraries importing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pandas.plotting import andrews_curves, parallel_coordinates, radviz
%matplotlib inline

# seaborn style
sns.set(style="whitegrid", color_codes=True)

# loading the dataset from UCI
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm", "Species"]

# reading dataset into Pandas DataFrame
df = pd.read_csv(url, header=None, names=columns)

# showing the first few rows
df.head()
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df, kind="scatter")
```

```python
g = sns.FacetGrid(df, hue="Species", height=5)
g.map(plt.scatter, "SepalLengthCm", "SepalWidthCm")
g.add_legend()
```

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=df)
```

```python
sns.violinplot(x="Species", y="PetalLengthCm", data=df)
```

```python
g = sns.FacetGrid(df, hue="Species", height=6)
g.map(sns.kdeplot, "PetalLengthCm", fill=True)
g.add_legend()
```

```python
sns.pairplot(df, hue="Species", height=3)
```

```python
df.boxplot(by="Species", figsize=(12, 8))
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 5))
andrews_curves(df, "Species")
plt.show()
```

## 📊 Iris Dataset Visualization Analysis

### 1. **Joint Plot (Sepal Length vs Sepal Width)**
- **Purpose:** Visualize the relationship between two numerical features.
- **Conclusion:** Shows that Setosa species is well-separated from the others based on sepal measurements, while Versicolor and Virginica have some overlap.

---

### 2. **FacetGrid Scatter Plot by Species**
- **Purpose:** Overlay scatter plots by species using color to show clustering.
- **Conclusion:** Reinforces that Setosa is linearly separable, while Versicolor and Virginica data points cluster closer, suggesting less clear separability.

---

### 3. **Box Plot (Petal Length by Species)**
- **Purpose:** Summarize the distribution of petal lengths for each species.
- **Conclusion:** Setosa has significantly smaller petals. Virginica shows the largest petal lengths with less overlap, making petal length a strong classification feature.

---

### 4. **Violin Plot (Petal Length by Species)**
- **Purpose:** Combines boxplot and KDE to show data distribution.
- **Conclusion:** Confirms that petal lengths are tightly grouped for Setosa, and Virginica has a wider, higher-range distribution—again showing petal length is a good distinguishing feature.

---

### 5. **KDE Plot (Petal Length by Species)**
- **Purpose:** Estimate probability densities for petal length by species.
- **Conclusion:** Shows non-overlapping, peaked distributions for Setosa, and some overlap between Versicolor and Virginica—supporting classification boundaries.

---

### 6. **Pairplot (All Pairwise Feature Relationships)**
- **Purpose:** Explore pairwise feature relationships and class separability.
- **Conclusion:** Petal measurements (length and width) give the clearest separation between species. Sepal features alone are not as effective.

---

### 7. **Boxplots Grouped by Species (All Features)**
- **Purpose:** Analyze feature value ranges across species.
- **Conclusion:** Most distinguishing features are petal-related. Setosa consistently shows lower values for both sepal and petal features.

---

### 8. **Andrews Curves**
- **Purpose:** Visualize high-dimensional data to detect patterns and outliers.
- **Conclusion:** Setosa clearly forms a distinct pattern, while Versicolor and Virginica overlap—supporting the idea that Setosa is linearly separable, but the others require more complex classification.

---

### 🧠 Final Insight:
**Petal Length and Petal Width are the most informative features** for classifying iris species. Setosa is linearly separable from the others, while Versicolor and Virginica require more advanced techniques like SVM or decision trees for accurate classification.
